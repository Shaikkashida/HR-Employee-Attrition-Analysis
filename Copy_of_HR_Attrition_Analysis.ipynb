```python
df['Attrition_Flag']=df['Attrition'].map({'Yes':1,'No':0})
```

```python
df.drop('Attrition_flag',axis=1,inplace=True)
```

```python
df['Age_Group']=pd.cut(
    df['Age'],
    bins=[18,25,35,45,55,65],
    labels=['18-25','26-35','36-45','46-55','56+'])
```

```python
df['Salary_slab']=pd.cut(
    df['MonthlyIncome'],
    bins=[0,3000,6000,10000,20000],
    labels=['low','medium','high','very high']
)
```

SQL ANALYSIS

```python
total_employees=df.shape[0]
attrition_count=df['Attrition_Flag'].sum()
attrition_rate=(attrition_count/total_employees)*100
total_employees,attrition_count,attrition_rate
```

```python
dept_attrition=df.groupby('Department')['Attrition_Flag'].mean()*100
dept_attrition.sort_values(ascending=False)
```

```python
role_attrition=df.groupby('JobRole')['Attrition_Flag'].mean()*100
role_attrition.sort_values(ascending=False)
```

```python
salary_attrition=df.groupby('Salary_slab')['Attrition_Flag'].mean()*100
salary_attrition
```

```python
Experience_attrition=df.groupby('YearsAtCompany')['Attrition_Flag'].mean()*100
Experience_attrition.head(6)
```
